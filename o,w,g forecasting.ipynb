```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('production data1.xlsx')
```

```python
df.info()
```

```python
df[['AVG_DOWNHOLE_PRESSURE',
    'AVG_DOWNHOLE_TEMPERATURE',
    'AVG_DP_TUBING',
    'AVG_ANNULUS_PRESS',
    'AVG_WHP_P',
    'AVG_WHT_P']] = df[['AVG_DOWNHOLE_PRESSURE',
                        'AVG_DOWNHOLE_TEMPERATURE',
                        'AVG_DP_TUBING',
                        'AVG_ANNULUS_PRESS',
                        'AVG_WHP_P',
                        'AVG_WHT_P']].ffill()
```

```python
df[['DP_CHOKE_SIZE','AVG_CHOKE_SIZE_P']] = df[['DP_CHOKE_SIZE','AVG_CHOKE_SIZE_P']].fillna(df[['DP_CHOKE_SIZE','AVG_CHOKE_SIZE_P']].median())
```

```python
df.info()
```

```python
df.head()
```

```python
df2=df.loc[1:6000]
df3=df.loc[6001:8933]
```

```python
df2
```

```python
X=df2[['AVG_DOWNHOLE_PRESSURE','AVG_DOWNHOLE_TEMPERATURE','AVG_DP_TUBING','AVG_CHOKE_SIZE_P']]
y=df2[['BORE_OIL_VOL','BORE_GAS_VOL','BORE_WAT_VOL']]
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=16)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
multi_output = MultiOutputRegressor(model)
multi_output.fit(X_train,y_train)
```

```python
y_pred=multi_output.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,accuracy_score,r2_score
```

```python
val=np.sqrt(mean_squared_error(y_test,y_pred))
val
mean_absolute_error(y_test,y_pred,multioutput='raw_values')
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
model2=RandomForestRegressor()
multi_output_clf1 = MultiOutputRegressor(model2)

multi_output_clf1.fit(X_train,y_train)
```

```python
y_pred2=multi_output_clf1.predict(X_test)
```

```python
r2_score(y_test,y_pred2)
```

```python
mean_absolute_error(y_test,y_pred2,multioutput='raw_values')
```

```python
importances = [regressor.feature_importances_ for regressor in multi_output_clf1.estimators_]

# Convert to DataFrame for better readability
importance_df = pd.DataFrame(importances, columns=X.columns, index=[f'Target {i+1}' for i in range(len(importances))])
print(importance_df)
```

```python
average_importance = importance_df.mean(axis=0)
print("Average Feature Importance:\n", average_importance)
plt.figure(figsize=(10, 6))
average_importance.sort_values().plot(kind='barh')
plt.title('Average Feature Importance')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model3=GradientBoostingRegressor()
multi_output_clf2 = MultiOutputRegressor(model3)

multi_output_clf2.fit(X_train,y_train)
```

```python
y_pred3=multi_output_clf2.predict(X_test)
```

```python
r2_score(y_test,y_pred3)
```

```python
mean_absolute_error(y_test,y_pred3,multioutput='raw_values')
```

```python
importances = [regressor.feature_importances_ for regressor in multi_output_clf2.estimators_]

# Convert to DataFrame for better readability
importance_df = pd.DataFrame(importances, columns=X.columns, index=[f'Target {i+1}' for i in range(len(importances))])
print(importance_df)
```

```python
average_importance = importance_df.mean(axis=0)
print("Average Feature Importance:\n", average_importance)
plt.figure(figsize=(10, 6))
average_importance.sort_values().plot(kind='barh')
plt.title('Average Feature Importance')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```
